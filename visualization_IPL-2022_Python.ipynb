Q2. Which team won the most matches?

```python
match_win = df['match_winner'].value_counts()
# df.groupby(['team1','team2'])['match_winner'].count()
match_win
```

```python
# Index â†’ category names (e.g., teams)
# Values â†’ numeric counts (e.g., number of wins)
sns.barplot(x=match_win.values , y = match_win.index)

# color to each bar 
sns.barplot(x=match_win.values , y = match_win.index , palette="rainbow" )
plt.xlabel("Number of wins")
plt.ylabel("Teams")
plt.title("Most winning matches by teams")
# label to each bar
# text(x_position, y_position, text_to_display)
# ax.text(x, y, text) â€” Function Meaning
# The text() function syntax is:
# label = sns.barplot(x=match_win.values , y = match_win.index , palette="viridis" )
# for i, v in enumerate(match_win.values):
#     label.text(i, v, v, ha="left", va="center")
```

 Q7.Top Scorers
 

```python
df.head()
Scores = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(5)
Scores
```

```python
# Scores.plot(kind ='bar')
# for horizantal 
Scores.plot(
    kind='barh',
    color=['lightblue', 'peachpuff', 'lightgreen', 'lightcoral', 'plum']
)
plt.xlabel("Scores")
plt.ylabel("Players Name")
plt.title("Top Scorers by players")
```

Q8.  Best Bowling Figures

```python
# df.head()
# Extract the wickets
#Extract the wickets from 'best_bowling_figure' by using function on 0 index
df['highest_wickets']=df['best_bowling_figure'].apply(lambda x:x.split('--')[0])
# convert data type sting to int
df['highest_wickets']=df['highest_wickets'].astype(int)
# find best or higest wicket  bowlers name 
bowler = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(5)
bowler
```

```python
bowler.plot(kind='barh',
          color=['lightblue', 'peachpuff', 'lightgreen', 'lightcoral', 'plum'] 
           )
plt.xlabel("Number Of Wiket")
plt.ylabel("Players Name ")
plt.title("Best Bowling Figures")
```
